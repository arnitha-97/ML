```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]   #numeric data
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]   #set has numbers 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_predicted=model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
model.predict(digits.data[0:5])
```

```python
from sklearn.metrics import confusion_matrix
#confusion matrix helps in finding how correct our model is 
cm=confusion_matrix(Y_test,y_predicted)  #give the actual and the predicted ones 


```

```python
import seaborn as sb 
plt.figure(figsize=(10,7))
sb.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('truth')

#wherever there are zeroes ,that is correct predictions
```
